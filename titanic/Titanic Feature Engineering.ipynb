```python
def age_imputer(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 38
        elif Pclass == 2:
            return 29
        else:
            return 25
    else:
        return Age
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(age_imputer, axis=1)
```

```python
# Making a dictionary of all the titles and mapping it to a new title column
```

```python
Title_dictionary = {
    'Capt': "Officer",
    'Col': "Officer",
    'Dr': "Officer",
    'Major': "Officer",
    'Rev': "Officer",
    'Jonkheer': "Royalty",
    'Don': "Royalty",
    'Sir': "Royalty",
    'the Countess': "Royalty",
    'Lady': "Royalty",
    'Ms': "Mrs",
    'Mrs': "Mrs",
    'Mr': "Mr",
    'Mme': "Mrs",
    'Mlle': "Miss",
    'Miss': "Miss",
    'Master': "Master",
}
```

```python
def get_titles():
    df["Title"] = df["Name"].map(lambda name:name.split(',')[1].split('.')[0].strip())
    
    # Mapping Each title to title dictionary
    df["Title"] = df.Title.map(Title_dictionary)
    df.drop("Name", axis=1, inplace=True)
    return df
```

```python
df = get_titles()
```

```python
df.head(10)
```

```python
df.info()
```

```python
sex_dummy = pd.get_dummies(df["Sex"], drop_first=True)
p_class_dummy = pd.get_dummies(df["Pclass"], drop_first=True)
embark_dummy = pd.get_dummies(df["Embarked"], drop_first=True)
title_dummy = pd.get_dummies(df["Title"], drop_first=True)
    
# Combining the three new dataframes to the training data
df = pd.concat([df, sex_dummy, p_class_dummy, embark_dummy, title_dummy], axis=1)
# Dropping the unnessary columns
df.drop(["Sex", "Pclass", "Embarked", "Title"], axis=1, inplace=True)
```

```python
train = df
```

```python
train.head()
```

```python
# Dropping Passenger ID and Ticket 
train.drop( ["Ticket", "PassengerId"], axis=1, inplace=True)
```

```python
train.head()
```

```python
train.info()
```

```python
X = train.drop('Survived', axis=1).values
y = train['Survived'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_classifier = RandomForestClassifier(n_estimators=200, min_samples_leaf=3, max_features=0.5, n_jobs=-1)
random_classifier.fit(X_train, y_train)
```

```python
Prediction = random_classifier.predict(X_test)
```

```python
# Checking the Metrics for Logistic Regression Model
from sklearn.metrics import confusion_matrix, classification_report
cr = classification_report(y_test, Prediction)
cm = confusion_matrix(y_test, Prediction)
```
