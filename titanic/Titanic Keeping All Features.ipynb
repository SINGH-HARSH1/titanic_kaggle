## Processing Titles from Names 

```python
# Processing the Titles, Using Set instead of a list.
title = set()
for name in X_train["Name"]:
    title.add(name.split(',')[1].split('.')[0].strip())
```

```python
# Making a dictionary of all the titles and mapping it to a new title column
Title_dictionary = {
    'Capt': "Officer",
    'Col': "Officer",
    'Dr': "Officer",
    'Major': "Officer",
    'Rev': "Officer",
    'Jonkheer': "Royalty",
    'Don': "Royalty",
    'Sir': "Royalty",
    'the Countess': "Royalty",
    'Lady': "Royalty",
    'Ms': "Mrs",
    'Mrs': "Mrs",
    'Mr': "Mr",
    'Mme': "Mrs",
    'Mlle': "Miss",
    'Miss': "Miss",
    'Master': "Master",
}

def get_titles():
    X_train["Title"] = X_train["Name"].map(lambda name:name.split(',')[1].split('.')[0].strip())
    
    # Mapping Each title to title dictionary
    X_train["Title"] = X_train.Title.map(Title_dictionary)
    
    return X_train
```

```python
X_train = get_titles()
```

## Processing The Age Column

```python
# Number of Missing Values in the Age Column
print(X_train.Age.isnull().sum())
```

We have used mean to fill the missing values in the other scripts, Now we will use median based on Columns like SEX, TITLE, Pclass etc, to fill the missing values in the Age Column

```python
grp_train = X_train.groupby(['Sex', 'Pclass','Title','Fare'])
grp_train_med = grp_train.median()
grp_train_med = grp_train_med.reset_index()[['Sex', 'Pclass','Title', 'Fare' ,'Age']]
grp_train_med.head()
```

```python
def fill_age(row):
    condition = (
        (grp_train_med['Sex'] == row['Sex']) & 
        (grp_train_med['Title'] == row['Title']) & 
        (grp_train_med['Pclass'] == row['Pclass'])&
        (grp_train_med['Fare'] == row['Fare'])
    )
    if np.isnan(grp_train_med[condition]['Age'].values[0]):
        condition = (
            (grp_train_med['Sex'] == row['Sex']) &
            (grp_train_med['Pclass'] == row['Pclass'])
        )
    return grp_train_med[condition]['Age'].values[0]

def process_age():
    global X_train
    
    X_train['Age'] = X_train.apply(lambda row: fill_age(row) if np.isnan(row['Age']) else row['Age'], axis=1)
    
    return X_train
```

```python
X_train = process_age()
```
